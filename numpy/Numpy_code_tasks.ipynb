```python
import pandas as pd
import numpy as np
```

```python


#1.Create Arrays

arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

```python
#2.Array Operations

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print(a + b)
print(a * b)
```

```python
#3.Zeros and Ones

print(np.zeros((2, 3)))
print(np.ones((3, 3)))
```

```python

#4.Mean, Max, Min

data = np.array([10, 20, 30, 40])

print("Mean:", np.mean(data))
print("Max:", np.max(data))
print("Min:", np.min(data))
```
