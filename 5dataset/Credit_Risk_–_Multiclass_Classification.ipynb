Data Preprocessing

```python
df.drop(["Name","ID","Customer_ID",'SSN'],axis=1,inplace=True)
```

```python
df["Payment_of_Min_Amount"].unique()
```

Label enoding

```python
df['Credit_Score']=df['Credit_Score'].map({'Good': 0, 'Poor': 1, 'Standard': 2})
```

Visulization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=df["Credit_Score"])
plt.title("Distribution of Credit Score Classes")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols[::]:   # show first 6 numeric features
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

target_corr = corr["Credit_Score"].sort_values(ascending=False)

plt.figure(figsize=(8,6))
target_corr.plot(kind="bar")
plt.title("Correlation of Features with Credit_Score")
plt.ylabel("Correlation Value")
plt.show()
```

```python
plt.figure(figsize=(6,8))

target_corr = corr[["Credit_Score"]].sort_values(by="Credit_Score", ascending=False)

sns.heatmap(target_corr,
            annot=True,
            cmap="coolwarm")

plt.title("Feature Correlation with Credit_Score")
plt.show()
```
