Data Preprocessing

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(["Name","ID","Customer_ID",'SSN'],axis=1,inplace=True)
```

Encode Target

```python
df["income"] = df["income"].str.strip()
```

```python
df["income"] = df["income"].map({
    "<=50K": 0,
    ">50K": 1
})
```
